```python
k5 = Rep_Char1[['ReplayID','Is Winner','HeroID']]
k4 = pd.DataFrame(columns = ['ReplayID', 'hero1','hero2','hero3','hero4','hero5','isWin'])
k6 = pd.DataFrame(columns = ['ReplayID', 'hero1','hero2','hero3','hero4','hero5','isWin'])
k7 = pd.DataFrame(columns = ['ReplayID', 'hero1','hero2','hero3','hero4','hero5','isWin'])
k5
```

```python
for j in range(0,10000):
    tmp_0 = []
    tmp_1 = []
    repID = k5['ReplayID'].iloc[10*j]
    for i in range(10*j+0, 10*j+10):
        if k5['Is Winner'].iloc[i]==1:
            tmp_1.append(k5['HeroID'].iloc[i])
        if k5['Is Winner'].iloc[i]==0:
            tmp_0.append(k5['HeroID'].iloc[i])
    tmp_0.sort()
    tmp_1.sort()
    k4.loc[2*j+0]=1
    k4.loc[2*j+1]=1
    k4.iloc[2*j+0]['ReplayID']=repID
    k4.iloc[2*j+0]['hero1']=tmp_0[0]
    k4.iloc[2*j+0]['hero2']=tmp_0[1]
    k4.iloc[2*j+0]['hero3']=tmp_0[2]
    k4.iloc[2*j+0]['hero4']=tmp_0[3]
    k4.iloc[2*j+0]['hero5']=tmp_0[4]
    k4.iloc[2*j+0]['isWin']=0
    k4.iloc[2*j+1]['ReplayID']=repID
    k4.iloc[2*j+1]['hero1']=tmp_1[0]
    k4.iloc[2*j+1]['hero2']=tmp_1[1]
    k4.iloc[2*j+1]['hero3']=tmp_1[2]
    k4.iloc[2*j+1]['hero4']=tmp_1[3]
    k4.iloc[2*j+1]['hero5']=tmp_1[4]
    k4.iloc[2*j+1]['isWin']=1
k4
```
